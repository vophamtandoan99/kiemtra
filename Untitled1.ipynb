```
def getNewListA(listA):
    while 10 in listA:
        listA.remove(10)
    return listA

listA  = [1,10,20,5,60,30,10,9,10]

newListA = getNewListA(listA)
print(newListA)
```

```
list1 = ["A","B","C","D","E","F"]
list2 = [1,2,3,4,5,6]
dic = {}
for list1_ in list1:
    for list2_ in list2:
        dic[list1_] = list2_
        list2.remove(list2_)
        break  
print (dic)
```

```
person = {"name": "Quoc Nam", "age": 28, "salary": 8000, "city": "Tuy Hoa"}
keyToRemove = ["salary", "city"]

def delete(person, keyToRemove):
  for x in keyToRemove:
      if x in person:
          person.pop(x)
  print(person)
delete(person, keyToRemove)
```
